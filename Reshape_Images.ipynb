# Resize images to (96x96) / (160x160)

```python
# load requirements
from numpy import load, asarray, savez_compressed
import numpy as np
import cv2
```

```python
# define function to resize

def resize_imgs(path, size = 96):
    images = list()
    data = load(path)['arr_0']
    for img in data:
        img = cv2.resize(img, (size,size))
        images.append(img)
    print("data_loaded")
    return asarray(images)
```

```python
# resize each dataset and save new array

sub_data_test_96 = resize_imgs("/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/sub_data_test.npz", 96)
print(sub_data_test_96.shape)
savez_compressed('/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/sub_data_test_96.npz', sub_data_test_96)


sub_data_train_96 = resize_imgs("/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/sub_data_train.npz", 96)
print(sub_data_train_96.shape)
savez_compressed('/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/sub_data_train_96.npz', sub_data_train_96)


data_train_160 = resize_imgs('/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/data_train.npz', 160)
print(data_train_160.shape)
savez_compressed('/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/data_train_160.npz', data_train_160)


data_test_160 = resize_imgs('/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/data_test.npz', 160)
print(data_test_160.shape)
savez_compressed('/Users/andrintschan/SWITCHdrive/ComputerVision/ProjectWork/data/data_test_160.npz', data_test_160)
```
